# Data Visualization 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Import complete")
```

```python
# Importing data
df=pd.read_csv("Topic_Survey_Assignment.csv")
```

```python
#sorting values
df.sort_values(by="Very interested",axis=0,ascending=False, inplace=True)
df
```

```python
# converting nos to percentage of total 2233
per_df=(df[df.columns]/2233)*100
pd.options.display.float_format = "{:,.2f}".format
per_df
```

```python
# visualizing the bar graph, with different formatting conditions
fig=plt.figure()
ax1=fig.add_subplot(1,1,1)


per_df.plot(kind='bar', ax=ax1, figsize=(20,8),legend=True, color=['#5cb85c','#5bc0de','#d9534f'],fontsize=14, width = 0.8)


totals = []

# find the values and append to list
for i in ax1.patches:
    totals.append(i.get_height())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax1.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax1.text(i.get_x(), i.get_height()+1,str(round(i.get_height(), 2))+'%', fontsize=15, color='black')
# setting graph title
ax1.set_title('Percentages of respondents Interest in Data Science Areas', fontdict = {'fontsize':16})

#remove chart borders except bottom one
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['left'].set_visible(False)
```
